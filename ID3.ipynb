```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
data1=pd.read_csv("C:\ZZZZ-MINE\Manipal\Semester 5\DMPA\OC_Marker.csv")
data1.head()
```

```python
data2=pd.read_csv("C:\ZZZZ-MINE\Manipal\Semester 5\DMPA\OC_Genarel_Chem.csv")
data2.head()
```

```python
data3=pd.read_csv("C:\ZZZZ-MINE\Manipal\Semester 5\DMPA\OC_Blood_Routine.csv")
data3.head()
```

```python
oc_blood_routine =data3.drop(columns=['TYPE','TYPE.1'])
oc_general_chem = data2.drop(columns=['TYPE','TYPE.1'])
print(oc_blood_routine)
```

```python
data = pd.concat([data1.set_index(['Age']),
                           oc_blood_routine.set_index(['Age']),
                           oc_general_chem.set_index(['Age'])],
                          axis=1).reset_index()

# Print the resulting combined dataset
print(data)
```

```python
#DROPPING DEPENDENT COLUMNS
columns_t_drop = ['TYPE','TYPE.1']
x = data.drop(columns=columns_t_drop,axis = 1)
x.head()
```

```python
#CLEANING CHECK
x[x.isnull().any(axis=1)].head()
```

```python
y = data['TYPE']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33, random_state=42)
```

```python
z = DecisionTreeClassifier(max_leaf_nodes=7, random_state=0)
z.fit(x_train,y_train)
```

```python
y_predicted=z.predict(x_test)
y_predicted
```

```python
accuracy_score(y_test,y_predicted)*100
```

```python
print(confusion_matrix(y_test,y_predicted))
print(classification_report(y_test,y_predicted))
```
